```python
# Step 1: Upload your file
from google.colab import files
import pandas as pd

uploaded = files.upload()

# Step 2: Get uploaded filename
filename = list(uploaded.keys())[0]

# Step 3: Try reading the file
try:
    df = pd.read_csv(filename)
    print(f"‚úÖ Successfully loaded '{filename}'\n")
    print("üìÑ First few rows of the file:")
    display(df.head())

    print("\nüìä Columns available in the file:")
    print(df.columns.tolist())

    # Step 4: Check if it's the correct file for EDA
    required_cols = ['Survived', 'Pclass', 'Sex', 'Age', 'Fare']
    if all(col in df.columns for col in required_cols):
        print("\n‚úÖ All required columns found. Proceeding with EDA...")

        import seaborn as sns
        import matplotlib.pyplot as plt

        sns.set(style="whitegrid")
        plt.rcParams['figure.figsize'] = (10, 6)

        # Clean data
        df['Age'].fillna(df['Age'].median(), inplace=True)

        # Plot survival count
        sns.countplot(x='Survived', data=df)
        plt.title("Survival Count")
        plt.show()

        # Survival by Gender
        sns.countplot(x='Sex', hue='Survived', data=df)
        plt.title("Survival by Gender")
        plt.show()

        # Age distribution
        sns.histplot(df['Age'], kde=True)
        plt.title("Age Distribution")
        plt.show()

    else:
        print("\n‚ö†Ô∏è The uploaded file doesn't contain the required columns for full Titanic EDA.")
        print("Required columns for EDA:", required_cols)
        print("You can still view and explore the current file manually.")

except Exception as e:
    print(f"‚ùå Error loading file: {e}")
```
