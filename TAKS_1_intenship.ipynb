```python
# Step 1: Upload file manually
from google.colab import files
uploaded = files.upload()

# Step 2: Import required libraries
import pandas as pd
import matplotlib.pyplot as plt

# (OPTIONAL) If you are using a specific uploaded file:
# Example: Reading a CSV file after upload
# df = pd.read_csv('your_filename.csv')
# display(df.head())

# Step 3: Create data manually based on the image (India 2022 population by age)
age_groups = ['020 Years', '2164 Years', '65+ Years']
population_millions = [512, 807, 98]         # in millions
percentages = [36.1, 57.0, 6.9]              # percentages
colors = ['#FFEB3B', '#2196F3', '#E91E63']   # yellow, blue, pink

# Step 4: Plotting the bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(age_groups, population_millions, color=colors, edgecolor='black')

# Step 5: Add labels and percentage on bars
for bar, pct in zip(bars, percentages):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2,
             height + 10,
             f'{height} Mn\n({pct}%)',
             ha='center',
             fontsize=12,
             fontweight='bold')

# Step 6: Styling and labeling the chart
plt.title("India's Population Distribution by Age Group (2022)", fontsize=16, fontweight='bold')
plt.xlabel("Age Groups", fontsize=12)
plt.ylabel("Population (in Millions)", fontsize=12)
plt.ylim(0, max(population_millions) + 100)
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Step 7: Show the chart
plt.tight_layout()
plt.show()
```
