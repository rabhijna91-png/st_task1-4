```python
# 1. Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Load Sample Dataset (from GitHub)
url = "https://raw.githubusercontent.com/plotly/datasets/master/2011_us_ag_exports.csv"
df = pd.read_csv(url)

# 3. Basic Info
print("‚úÖ Data Loaded Successfully")
print("Shape:", df.shape)
print("\nüßæ Columns:\n", df.columns)
print("\nüîç Null Values:\n", df.isnull().sum())

# 4. Rename columns to simulate accident context
df.rename(columns={
    'code': 'State_Code',
    'total exports': 'Accident_Count',
    'beef': 'Injuries',
    'pork': 'Fatalities',
}, inplace=True)

# 5. Exploratory Data Analysis
plt.figure(figsize=(12, 6))
sns.barplot(x='Accident_Count', y='state', data=df.sort_values(by='Accident_Count', ascending=False))
plt.title("üî∫ Simulated Accident Count by State")
plt.xlabel("Accident Count")
plt.ylabel("State")
plt.grid(True)
plt.tight_layout()
plt.show()

# 6. Correlation Heatmap
plt.figure(figsize=(8, 5))
sns.heatmap(df[['Accident_Count', 'Injuries', 'Fatalities']].corr(), annot=True, cmap='coolwarm')
plt.title("üìä Correlation Heatmap")
plt.show()

# 7. Describe Statistics
print("\nüìå Statistical Summary:\n", df[['Accident_Count', 'Injuries', 'Fatalities']].describe())

```
